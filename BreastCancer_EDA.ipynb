```python
#new variable 'mean_features' is created based on the mean of 'radius_mean', 'texture_mean', and 'area_mean
selected_features = ['radius_mean', 'texture_mean', 'area_mean']

df['mean_features'] = df[selected_features].mean(axis=1)

print(df['mean_features'])
```

```python
mean_radius = df['radius_mean'].mean()
df['radius_discrete'] = (df['radius_mean'] > mean_radius).astype(int)

# Print the first few rows of the DataFrame to inspect the changes
print(df['radius_discrete'])
```

```python
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Creating a StandardScaler object
scaler = StandardScaler()

# Standardizing the selected columns
df[numerical_columns] = scaler.fit_transform(df[numerical_columns])

# Inspecting the first few rows of the standardized DataFrame
print(df.head())
```

```python

# Plotting a histogram for 'radius_mean'
plt.figure(figsize=(10, 6))
plt.hist(df['radius_mean'], bins=20, color='blue', alpha=0.7)
plt.title('Distribution of Radius Mean in Breast Cancer Dataset')
plt.xlabel('Radius Mean')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
sns.set(style="whitegrid")

# Create a box plot for 'radius_mean' by 'diagnosis'
plt.figure(figsize=(10, 6))
sns.boxplot(x='diagnosis', y='radius_mean', data=df, palette='Set2')
plt.title('Box Plot of Radius Mean by Diagnosis')
plt.xlabel('Diagnosis (Malignant or Benign)')
plt.ylabel('Radius Mean')
plt.show()
```

```python
sns.set(style="whitegrid")

# Create a box plot for 'area_mean' by 'diagnosis'
plt.figure(figsize=(10, 6))
sns.boxplot(x='diagnosis', y='area_mean', data=df, palette='Set2')
plt.title('Box Plot of Area Mean by Diagnosis')
plt.xlabel('Diagnosis (Malignant or Benign)')
plt.ylabel('Area Mean')
plt.show()
```

```python
sns.set(style="whitegrid")

# Create a scatter plot for 'radius_mean' and 'texture_mean'
plt.figure(figsize=(10, 6))
sns.scatterplot(x='radius_mean', y='texture_mean', hue='diagnosis', data=df, palette='Set1')
plt.title('Scatter Plot of Radius Mean vs. Texture Mean')
plt.xlabel('Radius Mean')
plt.ylabel('Texture Mean')
plt.legend(title='Diagnosis')
plt.show()
```
